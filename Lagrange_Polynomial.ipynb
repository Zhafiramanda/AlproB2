```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
def divided_diff(x, y):
 '''
 fungsi untuk menghitung tabel beda hingga
 '''
 n = len(y)
 coef = np.zeros([n, n])
 # Kolom pertama adalah y
 coef[:,0] = y
 
 for j in range(1,n):
   for i in range(n-j):
     coef[i][j] = \
     (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])
 
 return coef

def newton_poly(coef, x_data, x):
 """
 Mengevaluasi Polinomial newton pada x
 """
 n = len(x_data) - 1 
 p = coef[n]
 for k in range(1,n+1):
     p = coef[n-k] + (x -x_data[n-k])*p
 return p
```

```python
x = np.array([0.0, 1.0, 2.0, 3.0, 4.0])
y = np.array([1.0000, 0.5403, -0.4161, -0.9900, -0.6536])

# dapatkan coef beda hingga
a_s = divided_diff(x, y)[0, :]

# evaluasi pada titik-titik data baru
x_new = np.arange(0.0, 4.1, .1)
y_new = newton_poly(a_s, x, x_new)

nX = 2.5
nY = newton_poly(a_s, x, nX)
print('f(x) pada x = 2.5 adalah {}'.format(nY))

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo')
plt.plot(x_new, y_new)
plt.title('Newton Polynomial')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [321.0, 322.8, 324.2, 325.0]
y = [2.50651, 2.50893, 2.51081, 2.51188]
P1_coeff = [2.41862*10**8, -2.23948*10**6, 6912, -7.11111]
P2_coeff = [-5.78735*10**7, 536875, -1660.12, 1.71111]
P3_coeff = [1.17866*10**7, -109499, 339.08, -0.35]
P4_coeff = [-1.4781*10**7, 137429, -425.92, 0.44]

# get the polynomial function
P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)
P4 = poly.Polynomial(P4_coeff)

x_new = np.arange(0, 325.1, 1)

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, P1(x_new), 'b', label = 'P1')
plt.plot(x_new, P2(x_new), 'r', label = 'P2')
plt.plot(x_new, P3(x_new), 'g', label = 'P3')
plt.plot(x_new, P4(x_new), 'y', label = 'P4')

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title('Lagrange Basis Polynomials')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

```python
L = 2.50651*P1 + 2.50893*P2 + 2.51081*P3 + 2.51188*P4
print('L_koef = ',L)

fig = plt.figure(figsize = (10,8))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show() 
```
